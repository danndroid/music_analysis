## When t=1,  2Pi has completed a full cycle! 

```python
frequency = 4
phase = 0.1

t = np.linspace(0, 1, 100) # 100 SAMPLES PER SECOND
s = np.sin(t*2*np.pi)
sf = np.sin(frequency*t*2*np.pi)
sfp = np.sin((frequency*t-phase)*2*np.pi)
```

```python
plt.figure(figsize=(15, 5))

plt.plot(t, s, color='green', linewidth=2.0, linestyle='-', label='signal') # ALCANZA DOBLE DISTANCIA
plt.plot(t, sf, color='blue', linewidth=2.0, linestyle='-', label=f'{frequency}f-signal')
plt.plot(t, sfp, color='red', linewidth=2.0, linestyle='-.', label=f'{frequency}fp-signal')


plt.plot([0, np.max(s)],[0,0], color='blue', linewidth=2.0, linestyle='-', alpha=0.2)
plt.plot(0, 0, "or")
plt.plot(1.0, 0, "or")


plt.xlabel('Time (seconds)', fontsize=15)
plt.ylabel('Amplitude', fontsize=15)
plt.legend(fontsize=15)
plt.tight_layout()
plt.show()
```
